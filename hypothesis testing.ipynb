```python
a = [5.1, 5.8, 6.0, 5.4, 5.7]
mean = np.mean(a) 
print(mean)
t,p = stats.ttest_1samp(a,mean)
print(f"t-stat: {t}, p-value: {p}")
```

## 2 sample test

```python
g1 = [1,2,3,4,5]
g2 = [6,7,8,9,10]
t,p = stats.ttest_ind(g1,g2)
print(t,p)
```

```python
import scipy.stats as stats
group1 = [5.2, 5.8, 6.1, 5.3, 5.9]
group2 = [6.2, 6.5, 6.8, 6.3, 6.1]
t_stat, p_value = stats.ttest_ind(group1, group2)
print(f"t-statistic: {t_stat}, p-value: {p_value}")
```

```python
g1 = [5.2, 5.8, 6.1, 5.3, 5.9]
g2 = [6.2, 6.5, 6.8, 6.3, 6.1]
t,p = stats.ttest_ind(g1,g2)
print(t,p)
```

```python
sns.kdeplot(group1, shade=True, label='Group 1', color='blue')
sns.kdeplot(group2, shade=True, label='Group 2', color='orange')
plt.axvline(np.mean(group1), label=f'Group 1 Mean: {np.mean(group1):.2f}')
plt.axvline(np.mean(group2), label=f'Group 2 Mean: {np.mean(group2):.2f}')
plt.title(f"Two-Sample t-Test (t-stat: {t_stat:.3f}, p-value: {p_value:.3f})")
plt.xlabel('Data Values')
plt.ylabel('Density')
plt.legend()
plt.show()
```

## paired t test 

```python
b = [5.2, 5.7, 6.0, 5.4, 5.8]
a = [6.0, 6.2, 6.5, 5.9, 6.1]
t_stat, p_value = stats.ttest_rel(b, a)
plt.plot(b, label='Before', color='blue')
plt.plot(a, label='After', color='green')
for i in range(len(b)):
    plt.plot([i, i], [b[i], a[i]], color='gray')  
plt.title(f"Paired t-Test (t-stat: {t_stat:.4f}, p-value: {p_value:.4f})")
plt.xlabel('Samples')
plt.ylabel('Values')
plt.legend()
plt.show()
```
