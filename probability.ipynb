```python
import numpy as np
import matplotlib.pyplot as plt
sample_means = [np.mean(np.random.uniform(0, 1, 100)) for _ in range(1000)]
plt.hist(sample_means, bins=30, density=True)
plt.title("Central Limit Theorem Demonstration")
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

```python
from scipy.stats import norm
mean = 0 
std_dev = 1  
normal_dist = norm(mean, std_dev)
result = normal_dist.rvs(size=10)  
print(result)
```

```python
from scipy.stats import poisson
rate = 3  
poisson_dist = poisson(rate)
result = poisson_dist.rvs(size=10)  
print(result)
```

```python
from scipy.stats import binom
n = 2  
p = 0.5 
binom_dist = binom(n, p)
result = binom_dist.rvs(size=10)  
print(result)
```

```python
from scipy.stats import bernoulli
p = 0.5
bernoulli_dist = bernoulli(p)
result = bernoulli_dist.rvs(size=10)  
print(result)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import bernoulli
p = 0.5  
data = bernoulli.rvs(p, size=1000)  
plt.hist(data, bins=2, density=True, edgecolor='black', alpha=0.7)
plt.xticks([0, 1])
plt.title("Bernoulli Distribution")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.show()
```

```python
from scipy.stats import binom

n = 10  
p = 0.5 
x = np.arange(0, n+1)
pmf_binom = binom.pmf(x, n, p)
plt.bar(x, pmf_binom, alpha=0.7, edgecolor='black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

```python
from scipy.stats import poisson
rate = 3  
x = np.arange(0, 15)
pmf_poisson = poisson.pmf(x, rate)
plt.bar(x, pmf_poisson, alpha=0.7, edgecolor='black')
plt.title("Poisson Distribution (Î»=3)")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.show()
```

```python
from scipy.stats import norm

mean = 0
std_dev = 1
x = np.linspace(-4, 4, 100)
pdf_norm = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf_norm, label="Normal Distribution", color="blue")
plt.fill_between(x, pdf_norm, alpha=0.3, color='blue')
plt.title("Normal Distribution (Î¼=0, Ïƒ=1)")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.show()
```

```python
from scipy.stats import norm

mean = 0
std_dev = 1
x = np.linspace(-4, 4, 100)
cdf_norm = norm.cdf(x, mean, std_dev)
plt.plot(x, cdf_norm, label="CDF", color="green")
plt.title("Cumulative Distribution Function (Normal Distribution)")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
# Load the Titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
probability_survival = titanic['survived'].mean()
print(f"Probability of Survival: {probability_survival:.2f}")
```

```python
from scipy.stats import bernoulli
p_survival = probability_survival 
bernoulli_dist = bernoulli(p_survival)
prob_survived = bernoulli_dist.pmf(1) 
prob_not_survived = bernoulli_dist.pmf(0) 
print(f"Probability of Surviving: {prob_survived:.2f}")
print(f"Probability of Not Surviving: {prob_not_survived:.2f}")
```
